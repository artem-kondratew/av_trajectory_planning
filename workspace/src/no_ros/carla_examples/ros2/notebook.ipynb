```python
import carla
```

```python
client = carla.Client('localhost', 2000)
```

```python
world = client.get_world()
map = world.get_map()
```

```python
for actor in world.get_actors():
    if 'audi' in actor.type_id:
        vehicle = actor
vehicle
```

```python
transform = vehicle.get_transform()
vehicle_location = transform.location
```

```python
current_lane = map.get_waypoint(vehicle_location, project_to_road=True)
lane_center = current_lane.transform.location
# lane_forward = waypoint.transform.get_forward_vector()

# dx = vehicle_location.x - lane_center.x
# dy = vehicle_location.y - lane_center.y

# normal_x = -lane_forward.y
# normal_y = lane_forward.x

# lateral_offset = dx * normal_x + dy * normal_y
# lateral_offset
current_lane.lane_id
```

```python
right_lane = current_lane.get_right_lane()
right_lane_center = right_lane.transform.location
right_lane_center.x, right_lane_center.y, right_lane_center.z
```
