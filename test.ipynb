```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ref = 10
v0 = 0
F = 2
Kp = 4
Ki = 0.8
Kd = 0.1
dt = 0.1
```

```python
def controller(Kp, Ki, Kd):
    v = v0
    t = 0

    varr = [v0]
    tarr = [t]

    I = 0
    e_prev = 0

    while t < 50:
        e = ref - v
        P = Kp * e
        I += Ki * e * dt
        D = Kd * (e - e_prev) / dt
        e_prev = e
        u = P + I + D
        v += (u - F) * dt
        varr.append(v)
        t += dt
        tarr.append(t)

    return tarr, varr
```

```python
tarr, pi = controller(Kp, Ki, 0)
tarr, pid = controller(Kp, Ki, Kd)

plt.plot(tarr, pi, color='b', label=r'$PI$')
plt.plot(tarr, pid, color='orange', label=r'$PID$')
plt.axhline(ref, color='r', linestyle='--')
plt.legend()
plt.show
```
